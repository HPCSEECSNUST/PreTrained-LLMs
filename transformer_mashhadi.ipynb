```python
from transformers import pipeline
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
# Load the model and tokenizer
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Create a text generation pipeline
text_generator = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
# Define prompts
prompts = [
    "Pakistan",
    "The greatest footballer ever",
    "The weather today",
    "One of the greatest scientific inventions",
]
```

```python
# Generate and print text for each prompt
for prompt in prompts:
    generated_text = text_generator(prompt, max_length=30, num_return_sequences=3)
    print(f"Prompt: {prompt}")
    print(f"Generated Text: 1. {generated_text[0]['generated_text']}\n 2. {generated_text[1]['generated_text']}\n 3. {generated_text[2]['generated_text']}\n")
```
